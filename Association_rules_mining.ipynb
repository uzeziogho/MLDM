```python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
#from mlxtend.frequent_patterns import apriori


from apyori import apriori
import utils
%matplotlib inline
```

```python
#Load dataset
obesity_data=pd.read_csv('/content/ObesityDataSet_raw_and_data_sinthetic.csv')
#dishonesty_data.columns=["CT","CU","LT","TC","TS"]
```

```python
#Analyse dataset
obesity_data.head(10)
```

```python
obesity_data = obesity_data.drop(columns=['Age','Height','Weight'])
```

```python
#creating sample transactions
transactions=[
    ['Female','yes','no','2.0','3.0','Sometimes','yes','1.0','no','2.0','Frequently','Motorbike','Normal_Weight'],
    ['Male','yes','no','1.0','2.0','Always','no','3.0','yes','1.0','Sometimes','Automobile','Insufficient_Weight'],
    ['Female','no','no','1.0','2.0','Never','yes','2.0','no','3.0','Sometimes','Walking','Obesity_Type_II'],
    ['Male','yes','yes','1.0','2.0','Sometimes','yes','1.0','yes','1.0','Always','Public_Transportation','Overweight_Level_I'],
    ['Male','yes','no','1.0','3.0','Always','no','2.0','yes','2.0','No','Public_Transportation','Obesity_Type_III'],
    ['Female','no','no','1.0','2.0','Sometimes','yes','3.0','yes','1.0','Sometimes','Automobile','Obesity_Type_I'],
    ['Male','yes','no','1.0','1.0','Sometimes','no','1.0','no','3.0','Sometimes','Public_Transportation','Overweight_Level_II'],
]
transactions
```

```python
# Generating association rules
Rules = list(apriori(transactions, min_support=0.5, min_confidence=0.25,min_lift=2, min_length=2))
```

```python
#Extract the created rules
#associationRules=utils.extract(Rules)
#rules_TR=pd.DataFrame(associationRules,columns=['LHS','RHS','Support','Confidence','Lift'])
#rules_TR
```

```python
for i in range(len(Rules)):
    LHS=list(Rules[i][2][0][0])
    RHS=list(Rules[i][2][0][1])
    support=Rules[i][1]
    confidence=Rules[i][2][0][2]
    lift=Rules[i][2][0][3]
    print("LHS:",LHS,"--","RHS:",RHS)
    print("Support:",support)
    print("Confidence:",confidence)
    print("Lift:",lift)
    print(10*"----")
```

```python
dte = TransactionEncoder()
dte_ary = dte.fit(obesity_data).transform(obesity_data)
disf = pd.DataFrame(dte_ary, columns=dte.columns_)
disf
```

```python
#Return item and itemsets with support of 60%
apriori(disf, min_support=0.00000000015,use_colnames=True)
```

```python
frequent_itemsets = apriori(disf, min_support=0.00045, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.8) ]
```
